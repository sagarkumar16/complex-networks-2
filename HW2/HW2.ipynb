```python
import matplotlib.pyplot as plt
import random
from HW2.HyperNetworkModel import HyperNetworkModel
```

# MAIN

```python
n = 10**3
gamma = 3
k = 10

hscm = HyperNetworkModel(n, k, gamma)
her = HyperNetworkModel(n, k, gamma)

hscm.create_ensemble("HSCM", 10)
her.create_ensemble("HER", 10**4)
```

```python
hscm_k, hscm_pk = hscm.degree_distribution('log', 25)
her_k, her_pk = her.degree_distribution('log', 25)
```

```python
plt.plot(hscm_k, hscm_pk, 'o', label="HSCM", color='blue', alpha=0.8)
plt.plot(her_k, her_pk, 'o', label="HER", color='orange', alpha=0.8)
plt.legend()
plt.ylabel(r'P(X=k)')
plt.xlabel('Degree (k)')
plt.loglog()
plt.title("Ensemble Degree Distributions")
plt.savefig("ensemble_degree_distributions.pdf")
```

```python
sample_hscm = hscm.ensemble[random.randint(0,9)]
sample_her = her.ensemble[random.randint(0, 10**4-1)]

sample_hscm_k, samples_hscm_pk = hscm.degree_distribution('log', 25, graph=sample_hscm)
sample_her_k, samples_her_pk = her.degree_distribution('linear', 25, graph=sample_her)

plt.plot(sample_hscm_k, samples_hscm_pk, 'o', label="HSCM")
plt.plot(sample_her_k, samples_her_pk, 'o', label="HER")
plt.legend()
plt.ylabel(r'P(X=k)')
plt.xlabel('Degree (k)')
plt.loglog()
plt.title("Sample Degree Distributions")
plt.savefig("sample_degree_distributions.pdf")
```
